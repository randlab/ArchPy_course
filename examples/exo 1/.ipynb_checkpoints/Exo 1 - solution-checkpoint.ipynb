# Exercice 1 - Simple 2D model

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import sys
import os
import geone
import geone.covModel as gcm
import pyvista as pv

sys.path.append("../../../../")
import ArchPy
```

## Building the model

### Grid

```python
# grid

sx = 0.5
sy = 1
sz = 0.15
x0 = 0
y0 = 0
z0 = -15
nx = 200
ny = 1
nz = 100

dimensions = (nx, ny, nz)
spacing = (sx, sy, sz)
origin = (x0, y0, z0)  
```

### Units and surfaces

```python
# units
#units covmodel
covmodelB = gcm.CovModel2D(elem=[('cubic', {'w':.5, 'r':[40, 40]})])  # To modify
covmodelA = gcm.CovModel2D(elem=[('cubic', {'w':7, 'r':[30, 30]})])


# create Units 
C = ArchPy.base.Unit(name = "C", order = 1, ID = 1, color = "lightblue", contact = "onlap",
                     surface = ArchPy.base.Surface(contact = "onlap"))


dic_s_B = {"int_method" : "grf_ineq","covmodel" : covmodelB, "mean" : -5}
surf_B = ArchPy.base.Surface(contact = "onlap", dic_surf = dic_s_B)
B = ArchPy.base.Unit(name = "B", order = 2, ID = 2, color="gold", contact = "onlap",
                     surface = surf_B)


dic_s_A = {"int_method" : "grf_ineq","covmodel" : covmodelA, "mean":-8}
surf_A = ArchPy.base.Surface(dic_surf=dic_s_A, contact="onlap")
A = ArchPy.base.Unit(name="A", order = 3,ID = 3, color="blue", contact = "onlap",
                     surface = surf_A)
```

### Pile

```python
P1 = ArchPy.base.Pile(name = "P1")
P1.add_unit([C, B, A])
```

### Arch table

```python
T1 = ArchPy.base.Arch_table(name = "Exo1", seed = 54)
T1.set_Pile_master(P1)  # define the main pile
T1.add_grid(dimensions, spacing, origin)  # grid
```

## Facies

```python
sand = ArchPy.base.Facies(ID = 1,name="Sand",color="yellow")
gravel = ArchPy.base.Facies(ID = 2,name="Gravel",color="lightgreen")
clay = ArchPy.base.Facies(ID = 4,name="Clay",color="blue")
silt = ArchPy.base.Facies(ID = 6,name="Silt",color="goldenrod")

C.add_facies([silt])
B.add_facies([sand, gravel])
A.add_facies([silt, clay])
```

### Modify pile

```python
covmodel_SIS_A = gcm.CovModel3D(elem = [("exponential", {"w":.25,"r":[70, 70, 5]})], alpha=0, name="vario_SIS") # input variogram
covmodel_SIS_B = gcm.CovModel3D(elem = [("exponential", {"w":.25,"r":[50, 50, 5]})], alpha=0, name="vario_SIS") # input variogram
lst_covmodelA = [covmodel_SIS_A] # list of covmodels to pass at the function
lst_covmodelB = [covmodel_SIS_B]

dic_f_B = {"f_method" : "SIS", "neig" : 20, "f_covmodel":covmodel_SIS_B}
dic_f_A = {"f_method" : "SIS","neig" : 20, "f_covmodel" : covmodel_SIS_A}

B.set_dic_facies(dic_f_B)
A.set_dic_facies(dic_f_A)
```

## Property

```python
# covmodels for the property model
covmodelK = gcm.CovModel3D(elem=[("exponential",{"w":0.3,"r":[5,5,1]})], name="K_vario")
covmodelK2 = gcm.CovModel3D(elem=[("spherical",{"w":0.1,"r":[3,3,1]})], name="K_vario_2")

permea = ArchPy.base.Prop("K",[sand, gravel, clay, silt],
                  [covmodelK2, covmodelK, covmodelK, covmodelK2],
                  means=[-3.5,-2,-8,-6.5],
                  int_method = ["sgs","sgs","sgs","sgs"],
                  def_mean=-5)
T1.add_prop([permea])
```

```python
T1.compute_prop(2)
```

```python
# retrieve facies data

plt.figure(figsize=(5, 7), dpi=150)
o = 1
for iu in range(2):
    for ifa in range(3):
        for ip in range(1):
            
            plt.subplot(4, 2, o)
            plt.imshow(T1.getprop("K", iu, ifa, ip, all_data=False)[:, 0], origin="lower")
            plt.title("un {} / fa {} / pr {}".format(iu, ifa, ip))
            
            o += 1
            
```
