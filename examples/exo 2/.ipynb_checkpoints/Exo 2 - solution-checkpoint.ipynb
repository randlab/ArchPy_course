# Exo 2 - Use MPS with ArchPy

This notebook shows some MPS capabilities of ArchPy using DeeSse, a valid license of Deesse is required.

The goal is to build a relatively complex, but simplified, 3D sedimentary model, inspired from the Herten site <https://www.sciencedirect.com/science/article/pii/S0022169411002174#b0005>.

The geological concept consists of 4 units (from younger to older):
- unit D --> homogeneous layer of silts
- unit C --> fluvio-glaclial deposits composed of open-gravels, sands and a silt-matrix with variable orientations
- unit B --> same as unit C
- unit A --> homogeneous layer of clay

We will simulate all the surfaces with MPS using a Training image (TI) of a braided river system and the unit B and C (fluvio-glacial deposits) are filled with MPS and a simple TI of ellipses. We will impose constraints on the orientations (azimuth and dip) of these deposits to mimick the depositional process.


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import sys
import os
import geone
import geone.covModel as gcm
import geone.deesseinterface as dsi
import pyvista as pv
import ArchPy
from ArchPy.base import Unit, Pile, Prop, Facies, Surface
```

```python
T1 = ArchPy.base.Arch_table(name = "Exo2", seed=10, working_directory = "ws_article")
```

```python
T1.add_grid(dimensions, spacing, origin)
```

### Stratigraphic pile

This is the simplified pile that have to be constructed


Concerning the simulation of the surfaces, to fit the values of the TI and the data, we will the use the ``"rescalingMode":"min_max"`` which implies to define a ``TargetMin`` and ``TargetMax`` for each surface. The values to put are above. Remainder: all these entries have to be put in the dictionnary of surface (``dic_surf`` argument in Surface).

Concerning the simulation of facies, we also add constraints on the orientation of the facies in unit B and C:
- Azimuth: follow the surfaces at bottom of the unit (``azi_bot``) and is fixed at 0° (towards N) at top (``azi_top``)
- Dip: follow the gradient at bottom (``dip_bot``) and fixed at 80° at top (``dip_top``)

To impose azimuth or dip to follow the surfaces you can just write "gradient" to the corresponding key of the dictionnary facies (``dic_facies``) of each unit.
For example in this example we will have for the azimuth at the bottom ``"azi_bot":"gradient"``.

```python

#create Units 

# Unit D
dic_f_D = {"f_method" : "homogenous"}  # DeeSse parameters
D = Unit(name="D",order=1, ID = 1, color="gold", contact="onlap",
         surface=Surface(contact="onlap"),dic_facies=dic_f_D)

# Unit C
dic_s_C_mps = {"int_method" : "MPS", "TI" : TI_topo,"TargetMin":-12, "TargetMax": 2, "rescalingMode":"min_max",
          "thresh":0.01,"maxscan":0.1}
dic_f_C = {"f_method" : "MPS", "TI":TI, "xr":1, "yr":1, "zr":.7, "maxscan":0.005, "thresh":0.01,
           "rot_usage":1,
           "rotAzi":"inference", "rotAziLoc":True, "azi_top":0, "azi_bot":"gradient",  # azimuth
           "rotDip":"inference", "rotDipLoc":True, "dip_top":80, "dip_bot":"gradient",  # dip
           "neig":12}  # DeeSse parameters

C = Unit(name="C",order=2, ID = 2, color="midnightblue", contact="onlap",
         dic_facies=dic_f_C, surface=Surface(dic_surf=dic_s_C_mps, contact="erode"))

# Unit B
dic_s_B_mps = {"int_method" : "MPS", "TI" : TI_topo,"TargetMin":-14, "TargetMax": 0, "rescalingMode":"min_max",
          "thresh":0.01,"maxscan":0.1}
B = Unit(name="B",order=3, ID = 3, color="green",contact="onlap",
         dic_facies=dic_f_C.copy(), surface=Surface(contact="erode", dic_surf=dic_s_B_mps))

# Unit A
dic_s_A_mps = {"int_method" : "MPS", "TI" : TI_topo,"TargetMin":-18, "TargetMax": -5, "rescalingMode":"min_max",
          "thresh":0.01,"maxscan":0.1}
dic_f_A = {"f_method" : "homogenous"} 
A = Unit(name="A",order=4, ID = 4, color="lightcoral",contact="onlap",
         dic_facies=dic_f_A, surface=Surface(dic_surf = dic_s_A_mps, contact="onlap"))
```

```python
#Master pile
P1 = Pile("P1")
P1.add_unit([D, C, B, A])

# set Pile master
T1.set_Pile_master(P1)
```

### Facies

You can look at the figure above to see the different facies that have to be created and where they need to go.

Before proceeding make sure:
- Facies name must match the one in the data (``.fd`` file). Make sure that they are identical. ArchPy is case sensitive.
- Facies ID match the one in the Training image. open-gravel is supposed to be below sand.


```python
# facies
f5 = Facies(5, "clay", "lightcoral")
f1 = Facies(1, "silt-matrix", "bisque")
f2 = Facies(3, "sand", "orangered")
f3 = Facies(2, "open-gravel", "royalblue")
f4 = Facies(4, "silts", "gold")

A.add_facies(f5)
B.add_facies([f1, f2, f3])
C.add_facies([f1, f2, f3])
D.add_facies(f4)
```

## Plots 

ArchPy offers multiple plotting functionnalities at different level such as :

``plot_unit(iu)``, ``plot_facies(iu, ifa)``, ``plot_prop(property, iu, ifa, ip)``, ``plot_proba (object)``, ``plot_mean(property, type = arithmetic)``, etc.

Slice can always be drawn with the slicex, slicey and slicez arguments. slices are drawn at fraction of the x, y and z axis. For example, slicez = 0.3 indicates to draw a slice along xy plane and cutting z at 0.3 of zmax.

here iu, ifa and ip are index for the different type of realizations made. iu = 1 for example indicating to plot the results related to the 2nd unit realization. Similarly, ifa and ip stand for the results associtated with the ifa facies and ip property realizations.

example. ``T1.plot_prop("K", iu=5, ifa=1, ip=0)`` indicates to plot the 1st property realization made on the 2nd facies simulation itself realized on the 6th unit realization.

```python
T1.plot_units(iu=0)
```

```python
T1.plot_facies(iu=0, ifa=0)
```

We can also filter values to plot only certains units/facies

```python
T1.plot_facies(excludedVal=[D.list_facies[0].ID, A.list_facies[0].ID])
```

Examples with slices and boreholes
